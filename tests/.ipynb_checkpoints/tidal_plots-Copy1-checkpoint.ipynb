```python
def plot_moon_tidal_only(moon, xmax=1e6, save=False, ecc_range=None):
        
    res = h5py.File(moon+'.hdf5', 'r')
    
    ecc = res['ecc'][:]
    time_range = res['time'][:]
    
    plt.rcParams.update({'font.size': 20})
    fig, ax1 = plt.subplots(1, 1, figsize=[20, 3.75], tight_layout=True)
    
    ax1.plot(time_range, ecc)   
    ax1.set_ylabel('ecc')
    ax1.set_xlabel('t [yr]')
    ax1.set_title(full_name)
    ax1.set_xlim(0,xmax)
    ax1.tick_params(direction='in', length=6, width=2, top=True, right=True)
    
    if ecc_range != None:
        ax1.set_ylim(ecc_range)
        
    if save:
        file_name = moon+'_kozai_tidal.png'
        plt.savefig(file_name)
    
    plt.show()
    res.close()
    

xmax = 2e5 # Change this to plot over a larger timescale and show more of the data (maximum of 5e6)

plot_moon_tidal_only('io', xmax=xmax)
plot_moon_tidal_only('eu', xmax=xmax)
plot_moon_tidal_only('ga', xmax=xmax)
plot_moon_tidal_only('ca', xmax=xmax)
```

```python
def plot_moon_with_kozai(moon, path, xmax=1e6, inc_range=None, ecc_range=None, save=False, save_path='plots/'):

    if moon == 'io':
        full_name = 'Io'
    elif moon == 'eu':
        full_name = 'Europa'
    elif moon == 'ga':
        full_name = 'Ganymede'
    elif moon == 'ca':
        full_name = 'Callisto'
    
    res = h5py.File(path+moon+'.hdf5', 'r')
    
    ecc = res['ecc'][:]
    inc = res['inc'][:]
    time_range = res['time'][:]
    
    plt.rcParams.update({'font.size': 20})
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=[20, 7.5], tight_layout=True, sharex=True)
    
    ax1.plot(time_range, inc)   
    ax1.set_ylabel('Inc [deg]')
    #ax1.set_xlabel('t [yr]')
    ax1.set_title('Inclination and Eccentricity '+full_name)
    ax1.set_xlim(0,xmax)
    ax1.tick_params(direction='in', length=6, width=2, top=True, right=True)
    if inc_range != None:
        ax1.set_ylim(inc_range)
    
    ax2.plot(time_range, ecc)   
    ax2.set_ylabel('Ecc')
    ax2.set_xlabel('t [yr]')
    ax2.set_xlim(0,xmax)
    ax2.tick_params(direction='in', length=6, width=2, top=True, right=True)
    if ecc_range != None:
        ax2.set_ylim(ecc_range)
    
    if save:
        file_name = moon+'_kozai_tidal.png'
        plt.savefig(save_path+file_name)
    
    plt.show()
    res.close()
```

```python
#change this depending on the machine you are working on
path = '/home/maarten/Documents/SMA/SMA_project/amuse-project/alice_data/'
save_path = '/home/maarten/Documents/SMA/SMA_project/amuse-project/plots/'

#change this to show more of the data, maximum of 5e6
xmax=5e4

#to change the y-range, change ecc_range or inc_range to display what you want to see

plot_moon_with_kozai('io',path=path,save=False,xmax=xmax,inc_range=(79.4,80.2),ecc_range=(0.15,0.27),save_path=save_path)
plot_moon_with_kozai('eu',path=path,save=False,xmax=xmax,ecc_range=(0.12,0.27),save_path=save_path)
plot_moon_with_kozai('ga',path=path,save=False,xmax=xmax,inc_range=(57,61),save_path=save_path)
plot_moon_with_kozai('ca',path=path,save=False,xmax=xmax,inc_range=(20,80),ecc_range=(0,1),save_path=save_path)
```
